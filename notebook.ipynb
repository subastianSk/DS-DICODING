Pertanyaan 2 : Bagaimana performa penjualan platform E-Commerce kami seiring berjalannya waktu?

```python
monthly_df = all_data.resample(rule='M', on='order_approved_at').agg({
    "order_id": "nunique",
})
monthly_df.index = monthly_df.index.strftime('%B') #mengubah format order_approved_at menjadi Tahun-Bulan
monthly_df = monthly_df.reset_index()
monthly_df.rename(columns={
    "order_id": "order_count",
}, inplace=True)
monthly_df.head()
```

```python
monthly_df = monthly_df.sort_values('order_count').drop_duplicates('order_approved_at', keep='last')
monthly_df.head()
```

```python
monthly_df.sort_values(by='order_count')
```

```python
month_mapping = {
    "January": 1,
    "February": 2,
    "March": 3,
    "April": 4,
    "May": 5,
    "June": 6,
    "July": 7,
    "August": 8,
    "September": 9,
    "October": 10,
    "November": 11,
    "December": 12
}

monthly_df["month_numeric"] = monthly_df["order_approved_at"].map(month_mapping)
monthly_df = monthly_df.sort_values("month_numeric")
monthly_df = monthly_df.drop("month_numeric", axis=1)
```

```python
import matplotlib.pyplot as plt

# Data
months = monthly_df["order_approved_at"]
order_count = monthly_df["order_count"]

# Warna yang berbeda
colors = ["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728", "#9467bd", "#8c564b", "#e377c2", "#7f7f7f", "#bcbd22", "#17becf"]

# Buat plot
plt.figure(figsize=(12, 6))
plt.bar(months, order_count, color=colors, alpha=0.7)
plt.xlabel('Month')
plt.ylabel('Order Count')
plt.title('Monthly Order Count for 2018', fontsize=16)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)

# Tampilkan nilai di atas setiap batang
for i, count in enumerate(order_count):
    plt.text(i, count + 100, str(count), ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```

Dalam grafik batang di atas, terlihat penurunan yang signifikan dalam jumlah pesanan pada bulan September, diikuti oleh peningkatan yang signifikan pada bulan November. Kami menggunakan grafik ini untuk memvisualisasikan fluktuasi bulanan dalam jumlah pesanan dengan jelas.

Berdasarkan customer_state

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Data bystate_df
bystate_df = all_data.groupby(by="customer_state").customer_id.nunique().reset_index()
bystate_df.rename(columns={
    "customer_id": "customer_count"
}, inplace=True)

# State dengan pelanggan terbanyak
most_common_state = bystate_df.loc[bystate_df['customer_count'].idxmax(), 'customer_state']

# Mengurutkan berdasarkan jumlah pelanggan
bystate_df = bystate_df.sort_values(by='customer_count', ascending=False)

# Pengaturan gaya plot
sns.set(style="whitegrid")

# Membuat plot
plt.figure(figsize=(12, 6))
sns.barplot(x='customer_state',
            y='customer_count',
            data=bystate_df,
            palette=["#FFC107" if state == most_common_state else "#D3D3D3" for state in bystate_df['customer_state']]
            )

plt.title("Number of Customers by State", fontsize=15)
plt.xlabel("State")
plt.ylabel("Number of Customers")
plt.xticks(fontsize=10)
plt.show()
```

Pertanyaan 6 : Dimana letak geografis yang memiliki pelanggan terbanyak?

```python
other_state_geolocation = geo_df.groupby(['geolocation_zip_code_prefix'])['geolocation_state'].nunique().reset_index(name='count')
other_state_geolocation[other_state_geolocation['count']>= 2].shape
max_state = geo_df.groupby(['geolocation_zip_code_prefix','geolocation_state']).size().reset_index(name='count').drop_duplicates(subset = 'geolocation_zip_code_prefix').drop('count',axis=1)
geolocation_silver = geo_df.groupby(['geolocation_zip_code_prefix','geolocation_city','geolocation_state'])[['geolocation_lat','geolocation_lng']].median().reset_index()
geolocation_silver = geolocation_silver.merge(max_state,on=['geolocation_zip_code_prefix','geolocation_state'],how='inner')

customers_silver = customers_df.merge(geolocation_silver,left_on='customer_zip_code_prefix',right_on='geolocation_zip_code_prefix',how='inner')

customers_silver.head()
```

```python
def plot_brazil_map(data):
    brazil = mpimg.imread(urllib.request.urlopen('https://i.pinimg.com/originals/3a/0c/e1/3a0ce18b3c842748c255bc0aa445ad41.jpg'),'jpg')
    ax = data.plot(kind="scatter", x="geolocation_lng", y="geolocation_lat", figsize=(10,10), alpha=0.3,s=0.3,c='maroon')
    plt.axis('off')
    plt.imshow(brazil, extent=[-73.98283055, -33.8,-33.75116944,5.4])
    plt.show()
```

```python
# Menghitung jumlah kunjungan atau tampilan halaman produk
page_views = len(all_data)

# Menghitung jumlah pesanan yang berhasil dibuat
successful_orders = len(all_data[all_data['order_status'] == 'delivered'])

# Menghitung rasio konversi
conversion_rate = successful_orders / page_views * 100

# Membuat visualisasi
plt.figure(figsize=(8, 6))
plt.bar(['Successful Orders', 'Page Views'], [successful_orders, page_views])
plt.ylabel('Count')
plt.title('Conversion Rate: {:.2f}%'.format(conversion_rate))
plt.show()
```

Pertanyaan 8: Apakah ada tren dalam jenis produk yang paling sering dikembalikan atau dibatalkan?
